```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
df=pd.read_csv('/Users/jeong-yula/Downloads/open/train.csv')
```

```python
drop_index = df[df['주택소유상태'] == 'ANY'].index
df = df.drop(drop_index)
```

```python
def year_encoder(x):
    output = x.replace('years', '')
    output = output.replace('year', '')
    output = output.replace('+', '')
    output = output.replace('<', '')
    output = output.replace('Unknown', '-1').strip()
    output = int(output)

    return output

df['근로기간'] = df['근로기간'].apply(year_encoder)
```

```python
pivot_table_period =df.pivot_table(index='근로기간', columns='대출기간',values='대출금액')
pivot_table_period
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table_period, annot=True, fmt='g', cmap='YlGnBu')
plt.title('대출금액')
plt.xlabel('대출기간')
plt.ylabel('근로기간')
plt.show()
```
